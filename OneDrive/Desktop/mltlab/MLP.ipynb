multilayer perceptron is the simplest type of ANN(artificial neural network).Each MLP model has an input layer ,output layer and hidden layers.The hidden layers are activated by activation functions like Relu,sigmoid,tanh,identity functions. These perceptrons in the layers are higly connected and parallel, this accelerates the computation


```python
import pandas as pd
data=pd.read_csv('HR_comma_sep.csv')
data.head()
```

preprocessing:Label encoding,map the salary column value:low:0,medium:1 and high:2.This is label encoding

```python
#import label encoder
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
data['salary']=le.fit_transform(data['salary'])
```

```python
data.head()
```

```python
#split the dataset
X=data[['salary','satisfaction_level','last_evaluation','number_project','average_montly_hours','time_spend_company','Work_accident','promotion_last_5years']]
y=data['left']
```

```python
#train test split function
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=30)
```

build employee churn using MLPclassifier

```python
#import MLP classifier
from sklearn.neural_network import MLPClassifier

clf=MLPClassifier(hidden_layer_sizes=(6,5),random_state=30,verbose=True,learning_rate_init=0.01)
clf.fit(X_train,y_train)
```

verbose:is used to print progress messages to standard ouptut
hidden_layer_sizes=(6, 5) 6 hidden layers with 5 neurons
learning_rate:step size
random_state=the initial random values for weights and for bias

```python
#make prediciton
#and evaluate
y_pred=clf.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
