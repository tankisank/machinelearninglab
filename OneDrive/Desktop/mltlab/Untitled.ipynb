```python
y = df.iloc[:, 1]
y
```

```python
x = df.drop("Column2", axis=1)
x
```

## Naive Bayes

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(xtrain, ytrain)
```

```python
ypred = nb.predict(x)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y, ypred)
ac = accuracy_score(y, ypred)
cm, ac
```

## PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
xpca = pca.fit_transform(x, y)
xpca
```

```python
plt.scatter(xpca[:,0][ypred ==0], xpca[:, 1][ypred == 0], label = "Class 1")
plt.scatter(xpca[:,0][ypred ==1], xpca[:, 1][ypred == 1], label = "Class 2")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.legend()
plt.grid()
```

```python
xpcatrain, xpcatest, ypcatrain, ypcatest = train_test_split(xpca, y, test_size=0.2)
```

```python
nb = GaussianNB()
nb.fit(xpcatrain, ypcatrain)
```

```python
ypred = nb.predict(xpca)
cm = confusion_matrix(y, ypred)
ac = accuracy_score(y, ypred)
cm, ac
```

```python
plt.scatter(xpca[:,0][ypred ==0], xpca[:, 1][ypred == 0], label = "Class 1")
plt.scatter(xpca[:,0][ypred ==1], xpca[:, 1][ypred == 1], label = "Class 2")

plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.legend()
plt.grid()
```
